```python
import pandas as pd
import plotly.express as px
from dash import Dash, html, dcc, Input, Output, callback
```

```python
airline_data = pd.read_csv(r"https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/airline_data.csv",
            encoding = "ISO-8859-1",
            dtype={'Div1Airport': str, 
                   'Div1TailNum': str, 
                    'Div2Airport': str, 
                    'Div2TailNum': str})
```

```python
airline_data.groupby(["Reporting_Airline"], as_index=False)["Flights"].sum().nlargest(10,"Flights")
```

```python
app = Dash(__name__)
# Design dash app layout

app.layout = html.Div(
                    children = [
                        
                                html.H1("Airline Dashboard",
                                        style = {
                                                "textAlign":"center",
                                                "color":"orange",
                                                "font-size":40
                                                }
                                        ),
                                html.Div(
                                        children=["Input: ",
                                                dcc.Input(
                                                            id="input-yr", 
                                                            value = "2010",
                                                            type = 'number',
                                                            style = {'height':'50 px', 
                                                                    "font-size":35}
                                                         )
                                                 ],
                                        style={
                                                "font-size":40
                                              }
                                        ),
                                html.Br(),
                                html.Br(),
                                html.Div(dcc.Graph(id="bar-plot"))      
                                ]
                    )
@app.callback(
    Output(component_id='bar-plot', component_property='figure'),
    Input(component_id='input-yr', component_property='value')
    
)
def get_graph(entered_year):
    # selecting data
    df = airline_data[airline_data["Year"]==int(entered_year)]
    # Top 10 airlines carrier in terms of number of flights
    g1=df.groupby(["Reporting_Airline"], as_index=False)["Flights"].sum().nlargest(10,"Flights")
    # Plot the graph
    fig1 = px.bar(g1, x = "Reporting_Airline", y= "Flights", 
                    title = "Top 10 airlines carrier in year "+ str(entered_year) + " in terms of number of flights")
    fig1.update_layout()
    return fig1

if __name__ == "__main__":
    app.run_server(port=8002, host = "127.0.0.1", debug = True)
    #app.run(debug=True)


```

Callback using two inputs

```python
app = Dash(__name__)
# Design dash app layout

app.layout = html.Div(
                    children = [
                        
                                html.H1("Airline Dashboard",
                                        style = {
                                                "textAlign":"center",
                                                "color":"orange",
                                                "font-size":40
                                                }
                                        ),
                                html.Div(
                                        children=["Year: ",
                                                dcc.Input(
                                                            id="input-yr", 
                                                            value = "2010",
                                                            type = 'number',
                                                            style = {'height':'50 px', 
                                                                    "font-size":35}
                                                         )
                                                 ],
                                        style={
                                                "font-size":40
                                              }
                                        ),
                                html.Div(
                                        children = ["State Abbreviation: ", 
                                                    dcc.Input(
                                                                id= "input-ab",
                                                                value = "AL",
                                                                type = "text",
                                                                style={
                                                                        "height":"50 px",
                                                                        "font-size":35
                                                                      }
                                                            )
                                                    ],
                                        style={"font-size":40}
                                        ),
                                html.Br(),
                                html.Br(),
                                html.Div(dcc.Graph(id="bar-plot"))      
                                ]
                    )
@app.callback(
    Output(component_id='bar-plot', component_property='figure'),
    [Input(component_id='input-yr', component_property='value'),
    Input(component_id="input-ab", component_property="value")]
    
)
def get_graph(entered_year, entered_state):
    # selecting data
    df = airline_data[(airline_data["Year"]==int(entered_year)) & (airline_data["OriginState"]==entered_state)]
    # Top 10 airlines carrier in terms of number of flights
    g1=df.groupby(["Reporting_Airline"], as_index=False)["Flights"].sum().nlargest(10,"Flights")
    # Plot the graph
    fig1 = px.bar(g1, x = "Reporting_Airline", y= "Flights", 
                    title = "Top 10 airlines carrier in year "+ str(entered_year) + " in terms of number of flights")
    fig1.update_layout()
    return fig1

if __name__ == "__main__":
    app.run_server(port=8002, host = "127.0.0.1", debug = True)
    #app.run(debug=True)
```
